## Finding out the correlation

```python
corrmatrix = credit.corr() 
fig = plt.figure(figsize = (7, 7)) 
sns.heatmap(corrmatrix, vmax = .8,square = True)
plt.show() 
```

## Splitting the dataset

```python
X = credit.drop(['Class'], axis = 1) 
Y = credit["Class"] 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify =Y,random_state=42) 
```

```python
X_train
```

```python
Y_train.value_counts()
```

```python
Y_test.value_counts()
```

## Model Building using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(X_train,Y_train)
y_pred=model1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
acc_lr = accuracy_score(y_pred,Y_test)
prec = precision_score(Y_test, y_pred) 
print("The precision is {}".format(prec)) 
print('The accuracy score is {} '.format(acc_lr))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```

```python
plt.figure(figsize =(9, 9)) 
sns.heatmap(cm, annot = True, fmt ="d"); 
plt.title("Confusion matrix") 
plt.ylabel('True class') 
plt.xlabel('Predicted class') 
plt.show() 
```

## Model Building using Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rfc = RandomForestClassifier() 
rfc.fit(X_train, Y_train) 
predictions = rfc.predict(X_test) 
```

```python
acc_rfc = accuracy_score(predictions,Y_test)
prec_rfc = precision_score(Y_test, predictions) 
print("The precision is {}".format(prec_rfc)) 
print('The accuracy score is {} '.format(acc_rfc))
```

```python
cm_rfc =confusion_matrix(predictions,Y_test)
cm_rfc
```

```python
LABELS = ['Normal', 'Fraud'] 
plt.figure(figsize=(9,9))
sns.heatmap(cm_rfc,xticklabels = LABELS,  
            yticklabels = LABELS,annot=True,fmt='d',cmap='viridis')
plt.title(f"Accuracy Score is : {acc_rfc}")
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.show()
```
